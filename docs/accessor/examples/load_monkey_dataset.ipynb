```python
import matplotlib.pyplot as plt
from wandb_addons.accessor import load_dataset
```

```python
dataset_builder = load_dataset("geekyrakshit/artifact-accessor/monkey_species:v0", quiet=True)
train_dataset = dataset_builder.as_dataset('train')
class_names = dataset_builder.info.features["label"].names
```

```python
sample = next(iter(train_dataset))
plt.imshow(sample["image"].numpy())
label_name = class_names[sample["label"].numpy()]
plt.title(f"Label: {label_name}")
plt.show()
```
